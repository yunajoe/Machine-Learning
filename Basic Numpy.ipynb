```python
# The Difference Between Copy and View

# copy is a new array, and the view is just a view of the original array.

# The copy owns mthe data and any changes made to the copy will not affect original array, and any changes made to the original array will not affect the copy.

# The view does not own the data and any changes made to the view will affect the original array, and any changes made to the original array will affect the view
```

```python
arr = np.array([1, 2, 3, 4, 5])
arr
```

```python
copy = arr.copy()
copy
```

```python
arr[0] = 1000
```

```python
print(arr)
print(copy)
```

```python
arr = np.array([1, 2, 3, 4, 5])
view = arr.view()
```

```python
arr[0] = 42
```

```python
print(arr)
print(view)
```

```python
arr = np.array([1, 2, 3, 4, 5])
x = arr.view()
x[0] = 30000
```

```python
print(arr)
print(x)
```

# Numpy 배열에 축 추가하기 

##  np.newaxis

```python
arr = np.array([1,2,3,4])
arr
```

```python
arr.shape
```

```python
arr.ndim
```

```python
# ':'를 사용해서 기존 배열의 값을 indexing
# np.newaxis attribute를 사용해서 길이가 1인 새로운 축

new = arr[:,np.newaxis]
print(new)
print(new.shape)
print(new.ndim)
```

```python
new2 = arr[np.newaxis,:]
print(new2)
print(new2.shape)
print(new2.ndim)
```

# shape(3, 4) 인 2차원의 배열에 길이가 1인 새로운 축을 추가해서 3차원 배열을 만들어보겠으! 


```python
new2 = np.arange(12).reshape(3,4)
print(new2)
print(new2.shape)
print(new2.ndim)
```

```python
new3 = new2[:,:,np.newaxis]
print(new3)
print(new3.shape)
print(new3.ndim)
```

```python
new3 = new2[:,np.newaxis,:]
print(new3)
print(new3.shape)
print(new3.ndim)
```
