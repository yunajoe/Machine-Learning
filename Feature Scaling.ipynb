# 1) Min-Max scaler
- This Scaler shrinks the data within the range of -1 to 1
- We can set the range like [0,1] or [0,5] or [-1,1].

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.DataFrame({'WEIGHT': [15, 18, 12,10],
                   'PRICE': [1,3,2,5]},
                   index = ['Orange','Apple','Banana','Grape'])
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler
```

```python
# Scaled feature
scaler.fit_transform(df)
```

```python
# Scaled feature with df
df1 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])
df1
```

```python
ax = df.plot.scatter(x='WEIGHT', y='PRICE',color=['red','green','blue','yellow'], 
                     marker = '*',s=100, label='BREFORE SCALING')  # s는 market size
ax
```

```python
bx = df1.plot.scatter(x='WEIGHT', y='PRICE',color=['red','green','blue','yellow'], 
                     marker = '*',s=100, label='AFTER SCALING')  # s는 market size
bx
```

```python
df1.plot.scatter(x='WEIGHT', y='PRICE', color=['red','green','blue','yellow'],
                   marker = 'o',s=60,label='AFTER SCALING', ax = ax
                )
```

# 2) Standard Scaler

-  assumes data is normally distributed within each feature and scales them such that the distribution centered around 0, with a standard deviation of 1.

- if data is not normally distributed, this is not that good

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df
```

```python
scaler.fit_transform(df)
```

```python
df2 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])
df2
```

```python
ax = df.plot.scatter(x='WEIGHT', y='PRICE',color=['red','green','blue','yellow'], 
                     marker = '*',s=80, label='BREFORE SCALING')
```

```python
bx = df2.plot.scatter(x='WEIGHT', y='PRICE',color=['red','green','blue','yellow'], 
                     marker = '*',s=80, label='AFTER SCALING')
```

```python
df2.plot.scatter(x='WEIGHT', y='PRICE', color=['red','green','blue','yellow'],
                 marker = 'o',s=60,label='AFTER SCALING', ax = ax)
```
