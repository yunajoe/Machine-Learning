```python
# scikit_learn 의 붓꽃 데이터 셋을 가져오기 

from sklearn.datasets import load_iris
irisData = load_iris()

# load_iris() 함수가 리턴하는 객체는 Bunch 클래스 객체
# Bunch 클래스 객체는 파이썬의 dictionary 와 유사한객체로
# 키와값으로 구성되어 있다

# 훈련데이터와 테스트데이터를 나누기위한함수
# train_test_split 모듈에 있는  train_test_split()
# train_test_split 모듈은 sklearn.jodel_selection 

from sklearn.model_selection import train_test_split

# scikit-learn  에서 데이터는 대문자 X로 표시하고 레이블은 소문자 y로 표시한다 

X_train, X_test, y_train, y_test = train_test_split( 
    irisData["data"], irisData["target"],random_state=0)   # random_state는 유사난소 생성기 


# train_test_split()의 리턴 타입은 모두 numpy 배열 



```

```python
X_train.shape   # 75 프로의 훈련 데이터 
```

```python
X_test.shape   # 25프로의 테스트 데이터
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
## KNN (K Nearest Neighbors )  >> 분류 머신 중 하나 >> 새로운 데이터의 가장 가까운 k개의 이웃을찾는다. 
## KNN 을사용하기 위해서는 neighbors 모듈에 KNeighborsClassifer 함수를 사용한다 
## KNeighborsClassfier() 함수에 중에 중요한 매개 변수는 n_neighbors >> 이 매개 변수는 이웃의 갯수를 지정하는 매개변수 

from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=1)

# 모델을 만들기 위해서는 fit 메서드 사용 
# fit메서드의 리턴값은 knn객체를 리턴한다 

knn.fit(X_train, y_train)  #
```

```python
# 채집한 iris의 새로운 데이터(sample)이라고 가정하고 Numpy 배열로 특성값을 만든다
# sckitit-learn 에서는 항상 데이터가 2차원 배열(대문자) 일 것으로 예측/ 소문자는 vector
```

```python
X_Newdata = np.array([[5.1,2.9, 1,0.3]])
X_Newdata.shape  # 종류가 1나인 (4개의 꽃특성, petal,sepal width 길이)
```

```python
# knn객체의 predict() 메서드를 사용하여 예측할 수 있다
prediction = knn.predict(X_Newdata)
prediction   # 새로운 데이터, X 는 품종이 0 (Setosa) 이다.
```

```python
irisData['target_names']
```

```python
print(f"예측:{prediction}")
print(f"예측 품종의 이름: {irisData['target_names'][prediction]}")
```

```python
y_predict = knn.predict(X_test)
y_predict   # X _ test의 예측값..< 이 값을 y_ test와[레이블] 비교해본다> 
```

```python
# 정확도를 계산하기 위해서 
# 1) numpy 의 mean() 메서드 이용  [또는] 
# 2 ) knn객체의 score() 메서드를 사용해도 된다 

# numyp mean() 메서드 이용 >> np.mean(y_predict == y_test) # 정확도가 약 97% 
```

```python
# knn score() 메서드 이용     
knn.score(X_test, y_test)
```
